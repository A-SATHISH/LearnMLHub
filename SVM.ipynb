```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from sklearn import svm
import numpy as np
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
data.head()
```

```python
data.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
data[data.Embarked.isnull()]
```

```python
data.drop(['Cabin'],axis=1,inplace=True)
data["Age"]=data["Age"].fillna(data["Age"].median())
data["Embarked"]=data["Embarked"].fillna("C")
data.Sex=data.Sex.replace(['male','female'],[0,1])
data.Embarked=data.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
data.dropna(inplace=True)
```

```python
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
```

```python
test.drop(['Cabin'],axis=1,inplace=True)
test["Age"]=test["Age"].fillna(data["Age"].median())
test["Embarked"]=test["Embarked"].fillna("C")
test.Sex=test.Sex.replace(['male','female'],[0,1])
test.Embarked=test.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
test.dropna(inplace=True)
```

```python
features=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
```

```python
c=0.7
clf=svm.SVC(kernel='linear',C=c).fit(data[features],data['Survived'])#features->independent,survival->dependent
```

```python
pred=clf.predict(test[features])
```
