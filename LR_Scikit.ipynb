```python
import pandas as pd
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/logistic_regression/insurance_data.csv")
data
```

```python
data.head()
```

```python
data["bought_insurance"].value_counts()
```

```python
x=data.drop('bought_insurance',axis='columns')
x
```

```python
y=data.bought_insurance
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
print("Total",len(x))
print("Train size",len(x_train))
print("Test size",len(x_test))
```

```python
model.fit(x_train,y_train)#fit randomly takes 80% datas ,so y_predicted will vary
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
print("Test data score",model.score(x_test,y_test))#bias=error,variance=fluctuations
print("Train data score",model.score(x_train,y_train))
```
