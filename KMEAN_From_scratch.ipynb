```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def initialize_centroids(x,k):
  indices=np.random.choice(len(x),k,replace=False)#picking centroids randomly
  return x[indices]
```

```python
def assign_to_clusters(x,centroids):
  distances=np.linalg.norm(x[:,np.newaxis]-centroids,axis=2)
  return np.argmin(distances,axis=1)
```

```python
def update_centroid(x,labels,k):
  centroids=np.array([x[labels==i].mean(axis=0) for i in range(k)])#mean calculation
  return centroids#updating centroids by mean
```

```python
def k_means(x,k,max_iters=100):
  centroids=initialize_centroids(x,k)
  for _ in range(max_iters):
    labels=assign_to_clusters(x,centroids)
    new_centroids=update_centroid(x,labels,k)

    if np.all(centroids==new_centroids):
      break

    centroids=new_centroids

  return labels,centroids
```

```python
np.random.seed(42)
x=np.concatenate([np.random.normal(loc=i,scale=1,size=(50,2)) for i in range(5)])
```

```python
x
```

```python
k=5
labels,centroids=k_means(x,k)
```

```python
plt.scatter(x[:,0],x[:,1],c=labels,cmap="viridis")
plt.scatter(centroids[:,0],centroids[:,1],marker='x',s=200,c='red',label='Centroid')
plt.title('K-means Clustering')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```
