```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
import numpy as np
```

```python
df =pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
test_df=pd.read_csv("https://github.com/ezioauditore-tech/AI/raw/main/datasets/Titanic/test.csv")
PassangerId=test_df["PassengerId"]
```

```python
final_df=df[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
final_df=final_df.dropna()
final_df.head()
```

```python
final_df["Sex"].replace(to_replace=final_df["Sex"].unique(),value=[1,0])#categorical to numerical
```

```python
final_df=pd.get_dummies(final_df,drop_first=True)
final_df
```

```python
train_y=final_df["Survived"]#output
train_x=final_df[['Pclass','Sex_male','Age','SibSp','Parch','Fare','Embarked_Q','Embarked_S']]
```

```python
from sklearn.model_selection import train_test_split
train_data,val_data,train_target,val_target=train_test_split(train_x,train_y,train_size=0.8)
```

```python
model=GaussianNB()
model.fit(train_data,train_target)
val_pred=model.predict(val_data)
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score: {0:0.4f}'.format(accuracy_score(val_target,val_pred)*100)+"%")
```
