```python
data['buying'],class_name=pd.factorize(data['buying'])
data['maint'],class_name=pd.factorize(data['maint'])
data['lug_boot'],class_name=pd.factorize(data['lug_boot'])
data['safety'],class_name=pd.factorize(data['safety'])
data['doors'],class_name=pd.factorize(data['doors'])
data['persons'],class_name=pd.factorize(data['persons'])
data.head()
```

```python
x=data.iloc[:,0:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3,random_state=0)#the values will taken continously (1st-70) for train and test in random_state=0
```

```python
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)
```

```python
y_train_pred=dtree.predict(x_train)
```

```python
accuracy=metrics.accuracy_score(y_train,y_train_pred)
accuracy
```

```python
y_test_pred=dtree.predict(x_test)
```

```python
accuracy=metrics.accuracy_score(y_test,y_test_pred)
accuracy
```
